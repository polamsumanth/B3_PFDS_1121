```python
import pandas as pd
import numpy as np


days = pd.to_datetime(pd.Series(range(1, 31), name='Day'), unit='D', origin='2023-01-01').dt.strftime('%Y-%m-%d')
temperatures = np.random.uniform(10, 30, size=30)

dummy_data = pd.DataFrame({'Day': days, 'Temperature': temperatures})


dummy_data.to_csv('daily_temperature.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("daily_temperature.csv")
plt.figure(figsize=(8,6))
plt.plot(data["Day"], data["Temperature"], marker="o", linestyle="-")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.title("Daily Temperature Variation - Weather Dataset")
plt.grid(True)
plt.show()

hottest_day = data.loc[data["Temperature"].idxmax()]
coldest_day = data.loc[data["Temperature"].idxmin()]

print("Hottest Day:", hottest_day["Day"], "with", hottest_day["Temperature"], "°C")
print("Coldest Day:", coldest_day["Day"], "with", coldest_day["Temperature"], "°C")
```

```python
import pandas as pd

data = {'Quarter': ['Q1', 'Q2', 'Q3', 'Q4'],
        'Sales': [10000, 15000, 12000, 18000]}
sales_df = pd.DataFrame(data)
sales_df.to_csv('sales_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("sales_data.csv")
plt.figure(figsize=(8,5))
plt.plot(data["Quarter"], data["Sales"], marker="o", linestyle="-", label="Sales")
peak = data.loc[data["Sales"].idxmax()]
plt.scatter(peak["Quarter"], peak["Sales"], color="red", s=100, zorder=5, label="Peak Sales")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.title("Quarterly Sales Performance - Retail Sales Dataset")
plt.grid(True)
plt.legend()
plt.show()

print("Peak Sales occurred in", peak["Quarter"], "with sales of", peak["Sales"])
```
