**BAR GRAPH**

```python
import matplotlib.pyplot as plt

students = ["A","B","C","D","E"]
marks = [85,70,90,60,75]
plt.bar(students, marks, color = "blue")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Marks")
plt.show()
```

**SCATTER PLOT**

```python
import matplotlib.pyplot as plt

hours = [2,3,4,5,6,7,8]
scores = [50,55,65,75,70,80,90]
plt.scatter(hours, scores, color = "red")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Hours vs Scores")
plt.show()
```

**HISTOGRAM**

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y, marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.show()
```

**SEA BORN**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Boxplot of Total Bill Day by Day ")
plt.show()
```

**PLOTLY USING IRIS**

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 size='petal_length')
fig.update_layout(width=1000, height=500) # Correct way to set figure size
fig.show()
```

**PANDAS VISUALIZATION**

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {"Month": ["Jan", "Feb", "Mar", "Apr"],
        "Sales" : [200, 250, 300, 400]}
df = pd.DataFrame(data)
df.plot(x="Month", y="Sales", kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.show()
```
