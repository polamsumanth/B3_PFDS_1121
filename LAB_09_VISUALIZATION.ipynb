**HEATMAP**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = np.random.rand(6,6)
corr = pd.DataFrame(data, columns=list("ABCDEF")).corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidth=0.5)
plt.title("Heatmap Example - Correlation Matrix")
plt.show()
```

**SEABORN**

```python
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset("iris")
sns.pairplot(df,hue="species",diag_kind="kde",palette="Set1")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#load data
tips=sns.load_dataset("tips")
```

**BARPLOT**

```python
plt.figure(figsize=(8,6))
sns.barplot(x="day",y="total_bill",data=tips,palette="muted")
plt.title("Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```

**BOXPLOT**

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day",y="total_bill",data=tips,palette="Set1")
plt.title("Distribution of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```

**COUNTPLOT**

```python
plt.figure(figsize=(8,6))
sns.countplot(x="day",data=tips,palette="Set2")
plt.title("Number of Appointments by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Number of Appointments")
```

**VOILIN**

```python
plt.figure(figsize=(8,6))
sns.violinplot(x="day",y="total_bill",data=tips,palette="Set3")
plt.title("Distribution of Total Bill by Day")
plt.xlabel("Day of the Week")
plt.ylabel("Total Bill")
plt.show()
```
